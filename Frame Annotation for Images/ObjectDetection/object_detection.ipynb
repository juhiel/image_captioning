```python
from imageai.Detection import ObjectDetection

detector = ObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath("yolo.h5")
detector.loadModel()
```

```python
import glob
import json
from tqdm import tqdm
result = dict()
i = 0

for image in glob.glob('train2014/*.jpg'):
    if i < 101:
        detections = detector.detectObjectsFromImage(input_image=image, output_image_path="imagenew.jpg", minimum_percentage_probability=30)

        result[image] = detections
        i +=1
        print(i)
    
with open('data.json', 'w') as f:
    json.dump(result, f)
```
